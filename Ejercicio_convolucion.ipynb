```python
import numpy as np # importamos la libreria numpy
h = [7, 0.5, 8, -2, 0.3] #señal al impulso
x = [10, -8, 3.5, 6] #señal de entrada

h_reverse = h[::-1] # invertimos h[n] para que sea h[-k]

y = np.convolve(x, h_reverse)
print(y)
```
